##Crear la clase "Planta", con 3 atributos (nombre, tipoRiego, altura), 2 funciones leerDatos(), mostrarDatos() y 2 objetos (planta1, planta2).

![texto del vínculo](https://raw.githubusercontent.com/a25308051280060-pixel/ExaUII-Clase-planta-Ambar-0060/refs/heads/main/infoclaseplanta2.png.png)

```python
class Planta:
    def __init__(self, nombre, tipoRiego, altura):
        self.nombre = nombre
        self.tipoRiego = tipoRiego
        self.altura = altura

    def leerDatos(self):
        self.nombre = input("Ingrese el nombre de la planta: ")
        self.tipoRiego = input("Ingrese el tipo de riego (ej. goteo, manual): ")
        self.altura = input("Ingrese la altura de la planta: ")
        return self.nombre, self.tipoRiego, self.altura

    def mostrarDatos(self):
        print("Nombre: ", self.nombre)
        print("Tipo de Riego: ", self.tipoRiego)
        print("Altura: ", self.altura)
        return self.nombre, self.tipoRiego, self.altura

# Fin de clase Planta

print("--- Registro de Planta 1 ---")
# Creando objeto planta1 con valores iniciales vacíos
planta1 = Planta("", "", "")
planta1.leerDatos()
planta1.mostrarDatos()

print("-*-*-*-*-*-*-*-*-*-*-*")

print("--- Registro de Planta 2 ---")
# Creando objeto planta2 con valores iniciales vacíos
planta2 = Planta("", "", "")
planta2.leerDatos()
planta2.mostrarDatos()

print("-*-*-*-*-*-*-*-*-*-*-*")
print("Software desarrollado por Ambar Fierro NL = 19")
```
